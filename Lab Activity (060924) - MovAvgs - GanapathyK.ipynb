```python
import numpy as np
import pandas as pd
```

#### Find moving average for the list of integers [4,8,12,16,20,24,28,32,36,40] with a window size of 3

```python
# Global Initialization
integers_list = [4,8,12,16,20,24,28,32,36,40]
window_size = 3
mov_avg_list = []
val_window_inner = []
val_window_outer = []
display_df = pd.DataFrame()

# Appending 0s to moving averages list at the start
for i in range(window_size - 1):
    mov_avg_list.append('-')
    val_window_outer.append('-')

# Moving averages calculation - main code
for value in range(len(integers_list) - window_size + 1):
    sum_list = 0
    val_window_inner = []
    for val in range(value, window_size + value):
        sum_list = sum_list + integers_list[val]
        val_window_inner.append(integers_list[val])
    val_window_outer.append(val_window_inner)
    mov_avg_list.append(sum_list/window_size)
## print('Original list:', integers_list, '\nLength:', len(integers_list), '\nMoving averages list:', mov_avg_list, '\nLength:', len(mov_avg_list))

# Displaying original & moving averages in a single display
display_df['Original'] = integers_list
display_df['Calc_Values'] = val_window_outer
display_df['Moving_Avg'] = mov_avg_list
display_df.reset_index(drop=True, inplace=True)
display_df
```
